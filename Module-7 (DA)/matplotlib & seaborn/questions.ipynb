```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

iris = sns.load_dataset('iris')


print(iris.head())
```

A. Creating and customizing line charts using Matplotlib

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Line Chart of Sepal Length vs Sepal Width for IRIS Dataset', fontsize=16)
plt.xlabel('Sepal Length', fontsize=12)
plt.ylabel('Sepal Width', fontsize=12)
plt.grid(True)
```

B. Visualizing relationships between two or more variables using scatter plots 

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')

plt.title('Scatter Plot of Sepal Length vs Sepal Width for IRIS Dataset', fontsize=16)
plt.xlabel('Sepal Length', fontsize=12)
plt.ylabel('Sepal Width', fontsize=12)
plt.grid(True)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
```

C. Studying distributions of variables using histograms & bar charts to 

```python
sns.histplot(iris['sepal_length'], bins=10, kde=True,element='step')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

sns.histplot(iris['sepal_width'], bins=10, kde=True,element='step')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')

plt.tight_layout()
```

F. Plotting multiple Matplotlib and Seaborn charts in a grid 

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Scatter plot of Sepal Length vs. Sepal Width
axs[0, 0].scatter(iris['sepal_length'], iris['sepal_width'], color='blue')
axs[0, 0].set_title('Sepal Length vs Sepal Width')
axs[0, 0].set_xlabel('Sepal Length')
axs[0, 0].set_ylabel('Sepal Width')

# Plot 2: Histogram of Sepal Length
axs[0, 1].hist(iris['sepal_length'], bins=15, color='orange')
axs[0, 1].set_title('Histogram of Sepal Length')
axs[0, 1].set_xlabel('Sepal Length')
axs[0, 1].set_ylabel('Frequency')

# Plot 3: Boxplot of Petal Length by Species
sns.boxplot(x='species', y='petal_length', data=iris, ax=axs[1, 0], palette='pastel')
axs[1, 0].set_title('Boxplot of Petal Length by Species')
```
